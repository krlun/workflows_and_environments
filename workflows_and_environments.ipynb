The data analysis was performed using Python with the supporting [NumPy](https://numpy.org) and [matplotlib](https://matplotlib.org) libraries.

The first code cell contains the needed imports and the subsequent cells does the data analysis step by step.

```python
import numpy as np
import matplotlib.pyplot as plt
```

The data was collected from  *The Best Beamline Ever* over a time period of less than 12 parsecs (thus rivaling the previous performance of the infamous smuggler [Han Solo](https://www.imdb.com/title/tt0076759/characters/nm0000148)) where the independent variable is the Force and the dependent variable is the number of Wookies observed.

The data is loaded using `np.genfromtxt`. To get a sense of a reasonable model, the data is then plotted using `matplotlib.pyplot`:

```python
data_file = 'data.csv'
data = np.genfromtxt(data_file, delimiter=',', skip_header=1)
lightsabers = data[:, 0]
wookies = data[:, 1]
plt.scatter(lightsabers, wookies, c='r')
plt.xlabel('lightsabers')
plt.ylabel('Wookies')
plt.show()
```

From a visual inspection of the plot above, a linear model of the form
$$
Wookies = k\cdot lightsabers + c
$$
for some slope $k$ and constant $c$ seems reasonble:

```python
def fit(x, k, c):
    return k*x + c

k, c = np.polyfit(lightsabers, wookies, 1)
x = np.linspace(np.min(lightsabers), np.max(lightsabers), 100)
plt.scatter(lightsabers, wookies, c='r', label='Data')
plt.plot(x, fit(x, k, c), label='Linear fit')
plt.xlabel('lightsabers')
plt.ylabel('Wookies')
plt.legend()
plt.show()
```

A visual inspection of the plot above thus concludes that this was indeed the world's best linear regression, ever.
